```python
merge_df = result.merge(right = post_processing, how = 'left',on ='card_id')
merge_df.head()
```

```python
def predict_row(row):
    if row['ratio'] ==-33.219281:
        return  -33.219281
    elif row['ratio1'] ==-33.219281:
        return -33.219281
    elif row['ratio2'] ==-33.219281:
        return -33.219281
    elif row['ratio3'] ==-33.219281:
        return  -33.219281
    elif row['ratio4'] ==-33.219281:
        return -33.219281
    elif row['ratio5'] ==-33.219281:
        return -33.219281
    elif row['ratio6'] ==-33.219281:
        return -33.219281
    else:
        return 1
merge_df['predict'] = merge_df.apply (lambda row: predict_row(row),axis=1)
merge_df.head()
```

```python
magic_number = merge_df['target'].quantile(0.0005) # to be divided by half next time -> 0.00005
def predict_row(row):
    if row['predict'] ==-33.219281:
        if row['target'] < magic_number:
            return  0
        else:
            return 1
    else:
        return 1

merge_df['outlier'] = merge_df.apply (lambda row: predict_row(row),axis=1)
merge_df.head() 
```

```python
merge_df['outlier'].value_counts()
```

```python
def prepare(row):
    if row['outlier'] == 0:
        return row['predict']
    else:
        return row['target']
    
merge_df['target'] =merge_df.apply (lambda row: prepare(row),axis=1)
```

```python
merge_df[['card_id','target']].to_csv('submission_post_processing.csv',index= False )
```
